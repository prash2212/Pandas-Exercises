```python
jcpa['Symbol']='AAPL'
jcpm['Symbol']='MSFT'
jcpf['Symbol']='FCBK'
jcpg['Symbol']='GOOG'
jcpt['Symbol']='TSLA'
jcpn['Symbol']='NTFX'
```

```python
jcpall = pd.concat([jcpm,jcpa,jcpg,jcpf,jcpt,jcpn]); jcpall.reset_index(inplace=True)
```

```python
jcpall.loc[(jcpall['Symbol'] == 'AAPL'),'Close'] = jcpall.loc[(jcpall['Symbol'] == 'AAPL'),'Close'] + 10
jcpall.loc[(jcpall['Symbol'] == 'GOOG'),'Close'] = jcpall.loc[(jcpall['Symbol'] == 'GOOG'),'Close'] + 20
jcpall.loc[(jcpall['Symbol'] == 'FCBK'),'Close'] = jcpall.loc[(jcpall['Symbol'] == 'FCBK'),'Close'] + 30
jcpall.loc[(jcpall['Symbol'] == 'TSLA'),'Close'] = jcpall.loc[(jcpall['Symbol'] == 'TSLA'),'Close'] + 40
jcpall.loc[(jcpall['Symbol'] == 'NTFX'),'Close'] = jcpall.loc[(jcpall['Symbol'] == 'NTFX'),'Close'] + 50
```

```python
jcpall.drop(['index'],axis=1, inplace=True); 
```

```python
jcpgrp = jcpall.groupby('Symbol'); 
```
