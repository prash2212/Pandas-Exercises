```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
appldata = pd.read_excel('../../../Data/all_stocks.xls',sheetname='AAPL')
msftdata = pd.read_excel('../../../Data/all_stocks.xls',sheetname='MSFT')
appldata['Symbol'] = 'AAPL'
msftdata['Symbol']='MSFT'
```

```python
alldata = pd.concat([appldata[:100],msftdata[:100]]); alldata[:20]
```

```python
jcp = alldata[['Date','Symbol','Adj Close']]; jcp
```

```python
jcpivot = jcp.pivot('Date','Symbol','Adj Close'); jcpivot
```

```python
jcpnoNan = jcpivot[jcpivot.notnull().all(axis=1)]
```

```python
jcpnoNan['AAPL'].plot()
```

```python
jcpnoNan.plot()
```

```python
msftvol = alldata.Volume.loc[(alldata.Symbol == 'MSFT')];
```

```python
plt.bar(np.arange(len(msftvol)),msftvol)
```

```python
msftvol.plot.bar()
```
